```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/desenvolvedorjairomonassa/EngenheiroDadosDesafio2/main/dados_jogadores-1.csv",sep=';',encoding='latin-1')
```

```python
#conversão simples de data ano-mes-dia
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'])
```

```python
df.loc[df['num_vitorias'].isna()]
```

```python
df[['num_vitorias']] = df[['num_vitorias']].fillna(value=0)
```

```python
df
```

```python
#Qual país possui a maior média de vitórias? No geral (independente do jogo). 
df.groupby('pais')['pais','num_vitorias'].sum().sort_values(ascending=False, by='num_vitorias')
```

```python
#Qual país possui o maior número de derrotas no Jogo da velha? 
df.groupby('pais')['pais','num_derrotas'].sum().sort_values(ascending=False, by='num_derrotas')
```

```python
#Qual o jogo que possui o menor número de partidas? 
df.groupby('jogo')['jogo','total_partidas'].sum().sort_values(ascending=True, by='total_partidas')
```

```python
#Qual o jogo preferido entre as mulheres? 
df.loc[df['genero']=='Feminino'].groupby('jogo').count()
```

```python
#Qual o país com mais jogadores de Dama? 
df.loc[df['jogo']=='Dama'].groupby('pais')['pais'].count().sort_values(ascending=False)
```
